### Filtering Data

```python
df.loc[df["Type 1"] == "Grass"]
```

```python
# for multiple conditions SEOERATE THEM WITH PARANTHESIS ()
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
```

```python
# for multiple conditions SEOERATE THEM WITH PARANTHESIS ()
df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")]
```

```python
new_filtered_df = df[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & ( df['HP'] > 65)]
new_filtered_df
```

#### Even We Have Filtered Data Still index is same, to overcome it we can use "RESET_INDEX"

```python
new_filtered_df = new_filtered_df.reset_index()
new_filtered_df
```

####  We Still Have old indexes in Data in Colums index we can use "RESET_INDEX(DROP=TRUE)"

```python
new_filtered_df = df[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & ( df['HP'] > 65)]
new_filtered_df = new_filtered_df.reset_index(drop=True)
new_filtered_df
```

#### Reset_index inplace

```python
new_filtered_df = df[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & ( df['HP'] > 65)]
new_filtered_df.reset_index(drop=True, inplace=True)
new_filtered_df
```

#### str.contains

```python
# All the Name which contains "Mega"
df[ (df['Name'].str.contains('Mega'))]
```

```python
# All the Name which do not contains "Mega"
df[ ~(df['Name'].str.contains('Mega'))]
```

#### REGEX : Regular Expressions

```python
import re
# df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)] # ignore casing Fire
```

```python
# find all the Name in which "pi" is present
df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)] # ignore casing Fire
```

```python
# find all the Name which STARTS with "pi" is present
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)] # ignore casing Fire
```

### Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified_csv.csv')
df.groupby(['Type 1']).mean().sort_values(["Defense"], ascending=False) # mean == average
```

```python
df.groupby(['Type 1']).mean().sort_values(["Attack"], ascending=False) # mean == average
```

```python
df.groupby(['Type 1']).sum().sort_values(['Legendary'], ascending=False)
```

```python
df.groupby(['Type 1']).count()
```

```python
df['count'] = 1
# added this column to df
```

```python
df.groupby(["Type 1", 'Type 2']).count()['count']
```

### Working with large amounts of data    

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified_csv.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, results])
new_df
```
